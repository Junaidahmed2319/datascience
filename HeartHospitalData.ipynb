```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (5, 3)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/master/heart.csv')
```

```python
df.head()
```

```python
df.columns = ['Age', 'Sex', 'ChestPainType', 'RestBp', 'Cholestrol', 'FBS', 'RestECG', 'MaxHeartRate', 'ExcerciseAngina', 'Oldpeak',
              'ST_Slope', 'NumVesselsFluro', 'Thal', 'Target' ]
```

```python
df.head
```

```python
sns.countplot(data=df, x='Target')
plt.title('Heart Disease')
plt.xticks([0,1],['No Disease','Disease'])
plt.show()
```

```python
#Age Distribution By Target
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Target', bins=30)
plt.title('Age Distribution By Target')
plt.show()
```

```python
#Chestpain vs Target
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='ChestPainType', hue='Target')
plt.title('Chestpain vs Target')
plt.xticks(rotation=45)
plt.show()
```

```python
#Maxheartrate vs Target
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='MaxHeartRate', hue='Target')
plt.title('Maxheartrate vs Target')
plt.show()
```

```python
df.groupby('Target')['MaxHeartRate'].mean()
```

```python
#cholestrol vs Target
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Cholestrol', hue='Target')
plt.title('Cholestrol vs Target')
plt.show()
```

```python
#excerciseangina vs target
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='ExcerciseAngina', hue='Target')
plt.title('excerciseangina vs target')
plt.show()
```

```python
corr = df.corr()
```

```python
plt.figure(figsize=(10, 8))  # Optional: control figure size
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
