```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['total_bill'], color='Blue')
plt.title('Distribution of total bill')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['tip'], color='Blue')
plt.title('Distribution of tip')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='sex',y='tip', color='Red')
plt.title('Tips by Gender')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['day'], color='Blue')
plt.title('Distribution of visit')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=df,x='sex')
plt.title('visits by gender')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
df2= df.select_dtypes(include='number')
df2.head()
```

```python
plt.figure(figsize=(5, 4))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['tip_percentage'] = (df['tip'] / df['total_bill']) *100
df.head()
```

```python
df['tip_percent'] = df['tip_percentage'].round(2).astype(str) + '%'
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='day',y='tip_percentage', color='Red')
plt.title('Tips by Gender Day Wise')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='smoker',y='tip_percentage', color='Red')
plt.title('Tips by Smoker')
plt.show()
```

```python
df.groupby('smoker')['tip_percentage'].mean()
```

```python
df.groupby('time')['total_bill'].mean()
```

```python
sns.barplot(x='', y='Price', data=df)
plt.title('Average Price per Location')
plt.show()
```
